```python
import os

import pandas as pd
```

```python
path = '../data/football-data.co.uk'
files = os.listdir(path)
files
```

```python
dfs = []
for file in sorted(files):
    df = pd.read_csv(os.path.join(path, file))
    df["Season"] = file.split(".")[0]
    df["GameDay"] = df.index // 9 + 1
    dfs.append(df)

df_all = pd.concat(dfs, axis=0).reset_index(drop=True)
df_all
```

```python


df_all["Result"] = df_all.apply(lambda row: f"{row['FTHG']}:{row['FTAG']}", axis=1)
df_all[["Season", "GameDay","Date", "FTHG", "FTAG", "Result", "FTR", 'BWCH', 'BWCD', 'BWCA']]
```

```python
#Value counts of results normalized per season
result_counts = df_all.groupby("Season")["Result"].value_counts(normalize=True).unstack().transpose()
# keep only most common results
most_common_results = result_counts.sum(axis=1).sort_values(ascending=False).head(10).index
result_counts.loc[most_common_results]
```
